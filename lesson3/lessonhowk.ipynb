```python
import requests
import csv
from requests import Response
url = 'https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=5481753E-52AF-40DA-9A8A-9E192B245E13'
respones:Response = requests.get(url)
print(respones.text)
with open('a1.csv','w',encoding="utf-8",newline='') as f :
    f.write(respones.content.decode('utf-8'))
```

```python
import json
url = 'https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=44E49A04-8DE8-4E74-B872-BE322286C940'
response:Response = requests.get(url)
print(response.text)
with open('a2.json','w',encoding='utf-8',newline='') as ff:
    data = response.json()
    json.dump(data,ff,ensure_ascii=False,indent=4)
```
